```python
import json
import nltk
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import string
from sklearn.feature_extraction.text import TfidfVectorizer

# Ensure all NLTK resources are downloaded
try:
    nltk.download('punkt', quiet=True)
    nltk.download('stopwords', quiet=True)
    nltk.download('wordnet', quiet=True)
    nltk.download('omw-1.4', quiet=True)
except:
    print("NLTK downloads failed - using alternative methods")

# Initialize tools with fallbacks
stop_words = set(nltk.corpus.stopwords.words('english')) if 'stopwords' in nltk.data.find('corpora') else set()
stemmer = nltk.PorterStemmer()
lemmatizer = nltk.WordNetLemmatizer() if 'wordnet' in nltk.data.find('corpora') else None

def simple_tokenize(text):
    # Fallback tokenizer if punkt fails
    return text.lower().split()

def preprocess_text(text):
    try:
        # Try using NLTK's word_tokenize first
        try:
            tokens = nltk.word_tokenize(text.lower())
        except:
            tokens = simple_tokenize(text.lower())

        # Remove punctuation and stopwords
        tokens = [word for word in tokens if word not in stop_words and word not in string.punctuation]

        # Stemming
        stemmed = [stemmer.stem(word) for word in tokens]

        # Lemmatization if available
        if lemmatizer:
            processed = [lemmatizer.lemmatize(word) for word in stemmed]
        else:
            processed = stemmed

        return ' '.join(processed)
    except Exception as e:
        print(f"Error processing: '{text}' - using simple preprocessing")
        # Fallback - simple lowercase and split
        return ' '.join([word.lower() for word in text.split()
                        if word.lower() not in stop_words and word not in string.punctuation])

# Load the dataset
with open('intents 2.json') as f:
    data = json.load(f)

# Preprocess all patterns
preprocessed_data = []
tags = []
for intent in data['intents']:
    for pattern in intent['patterns']:
        processed = preprocess_text(pattern)
        if processed:  # Only add if we got something
            preprocessed_data.append(processed)
            tags.append(intent['tag'])

# Create DataFrame
df = pd.DataFrame({
    'text': preprocessed_data,
    'tag': tags
})

print(f"Successfully processed {len(df)} samples")
print("\nSample of preprocessed data:")
print(df.head())

# TF-IDF Vectorization
try:
    tfidf = TfidfVectorizer(max_features=1000)
    tfidf_vectors = tfidf.fit_transform(df['text'])
    print("\nTF-IDF Vectorization successful!")
except Exception as e:
    print(f"\nTF-IDF Vectorization failed: {e}")
    print("Using simple word counts instead")
    from sklearn.feature_extraction.text import CountVectorizer
    cv = CountVectorizer(max_features=1000)
    tfidf_vectors = cv.fit_transform(df['text'])

# Visualization 1: Word Cloud
if len(df) > 0:
    all_text = ' '.join(df['text'])
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('Most Common Words in Chatbot Patterns')
    plt.show()
else:
    print("No data available for word cloud")

# Visualization 2: Tag Distribution
if len(df) > 0:
    plt.figure(figsize=(12, 6))
    df['tag'].value_counts().plot(kind='bar')
    plt.title('Distribution of Intents/Tags')
    plt.xlabel('Intent Tag')
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
else:
    print("No data available for tag distribution")

print("\nPreprocessing completed!")
```

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import string
from sklearn.feature_extraction.text import TfidfVectorizer
import re
from collections import defaultdict

# Custom tokenizer that doesn't require NLTK
def simple_tokenizer(text):
    # Split on whitespace and remove punctuation
    return [word.strip(string.punctuation) for word in text.lower().split() if word.strip(string.punctuation)]

# Custom stop words list
basic_stopwords = {
    'i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', 'your', 'yours',
    'yourself', 'yourselves', 'he', 'him', 'his', 'himself', 'she', 'her', 'hers',
    'herself', 'it', 'its', 'itself', 'they', 'them', 'their', 'theirs', 'themselves',
    'what', 'which', 'who', 'whom', 'this', 'that', 'these', 'those', 'am', 'is', 'are',
    'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had', 'having', 'do', 'does',
    'did', 'doing', 'a', 'an', 'the', 'and', 'but', 'if', 'or', 'because', 'as', 'until',
    'while', 'of', 'at', 'by', 'for', 'with', 'about', 'against', 'between', 'into',
    'through', 'during', 'before', 'after', 'above', 'below', 'to', 'from', 'up', 'down',
    'in', 'out', 'on', 'off', 'over', 'under', 'again', 'further', 'then', 'once', 'here',
    'there', 'when', 'where', 'why', 'how', 'all', 'any', 'both', 'each', 'few', 'more',
    'most', 'other', 'some', 'such', 'no', 'nor', 'not', 'only', 'own', 'same', 'so',
    'than', 'too', 'very', 's', 't', 'can', 'will', 'just', 'don', 'should', 'now'
}

# Simple stemmer (no NLTK required)
def simple_stem(word):
    if word.endswith('ing'):
        return word[:-3]
    elif word.endswith('ly'):
        return word[:-2]
    elif word.endswith('s'):
        return word[:-1]
    return word

def analyze_text_processing(text):
    print(f"\nOriginal text: '{text}'")

    # Tokenization
    tokens = simple_tokenizer(text)
    print(f"Tokenized: {tokens}")

    # Remove stopwords
    filtered = [word for word in tokens if word not in basic_stopwords]
    print(f"After stopword removal: {filtered}")

    # Stemming
    stemmed = [simple_stem(word) for word in filtered]
    print(f"Stemmed: {stemmed}")

    # Lemmatization simulation (without WordNet)
    lemma_map = {
        'are': 'be', 'am': 'be', 'is': 'be', 'was': 'be', 'were': 'be',
        'running': 'run', 'ran': 'run', 'happier': 'happy', 'happiest': 'happy',
        'better': 'good', 'best': 'good'
    }
    lemmatized = [lemma_map.get(word, word) for word in filtered]
    print(f"Lemmatized: {lemmatized}")

    # Full preprocessing
    processed = ' '.join([lemma_map.get(simple_stem(word), simple_stem(word)) for word in filtered])
    return processed

# Load dataset
with open('intents 2.json') as f:
    data = json.load(f)

# Process sample texts to demonstrate each step
sample_texts = [
    "I'm feeling very anxious about my exams",
    "What is the difference between sadness and depression?",
    "Good evening, how has your day been?"
]

print("="*60)
print("INDIVIDUAL PROCESSING STEP EXAMPLES")
print("="*60)
for text in sample_texts:
    analyze_text_processing(text)
    print("-"*60)

# Preprocess all patterns
preprocessed_data = []
tags = []
original_texts = []
for intent in data['intents']:
    for pattern in intent['patterns']:
        original_texts.append(pattern)
        processed = analyze_text_processing(pattern)
        preprocessed_data.append(processed)
        tags.append(intent['tag'])

# Create DataFrame
df = pd.DataFrame({
    'original_text': original_texts,
    'processed_text': preprocessed_data,
    'tag': tags
})

# TF-IDF Vectorization
print("\n" + "="*60)
print("TF-IDF FEATURE EXTRACTION")
print("="*60)
tfidf = TfidfVectorizer(max_features=1000)
tfidf_vectors = tfidf.fit_transform(df['processed_text'])
print(f"Created TF-IDF matrix with shape: {tfidf_vectors.shape}")

# Visualization
plt.figure(figsize=(15, 10))

# Word Cloud
plt.subplot(2, 2, 1)
all_text = ' '.join(df['processed_text'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Processed Text')

# Tag Distribution
plt.subplot(2, 2, 2)
tag_counts = df['tag'].value_counts()
tag_counts.plot(kind='bar')
plt.title('Distribution of Intent Tags')
plt.xlabel('Intent Tag')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')

# TF-IDF Feature Importance
plt.subplot(2, 2, 3)
sum_tfidf = tfidf_vectors.sum(axis=0)
tfidf_scores = [(word, sum_tfidf[0, idx]) for word, idx in tfidf.vocabulary_.items()]
tfidf_scores = sorted(tfidf_scores, key=lambda x: x[1], reverse=True)[:15]
words = [x[0] for x in tfidf_scores]
scores = [x[1] for x in tfidf_scores]
plt.barh(words, scores)
plt.title('Top 15 Words by TF-IDF Score')
plt.xlabel('TF-IDF Score')
plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()

# Save processed data
df.to_csv('preprocessed_intents.csv', index=False)
print("\nSaved preprocessed data to 'preprocessed_intents.csv'")

# Summary
print("\n" + "="*60)
print("PREPROCESSING SUMMARY")
print("="*60)
print(f"Total samples processed: {len(df)}")
print(f"Unique intent tags: {len(df['tag'].unique())}")
print("\nFeature extraction methods applied:")
print(f"- TF-IDF vectors ({tfidf_vectors.shape[1]} features)")
print("\nVisualizations created:")
print("- Word cloud of most frequent terms")
print("- Bar chart of intent tag distribution")
print("- Horizontal bar chart of top TF-IDF features")
```
